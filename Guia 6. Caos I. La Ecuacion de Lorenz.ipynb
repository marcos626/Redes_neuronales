## Ejercicio 2)

Considere el sistema de Ecuaciones Diferenciales Ordinarias de Lorenz

\begin{eqnarray}
\frac{dx}{dt} & = & s(y-x) \\
\frac{dy}{dt} & = & rx-y-xz \\
\frac{dz}{dt} & = & xy-bz
\end{eqnarray}

Sea $p=(s,r,b)$ el vector de parámetros.

**2.1)**
Usando el método de Runge-Kutta de orden 4, resuelva la ODE para las siguientes elecciones de las condiciones iniciales, de los parámetros y del intervalo de tiempo de integración:

&nbsp; &nbsp; **i)** $(x(0),y(0),z(0))=(1,0.5,0.1)$, $p=(10,0.5,8/3)$ y $t\in [0,10]$.

```python
# 2.1)
def lorenz(x, t, p):
    s, r, b = p
    dxdt = s * (x[1] - x[0])
    dydt = r * x[0] - x[1] - x[0] * x[2]
    dzdt = x[0] * x[1] - b * x[2]
    return np.array([dxdt, dydt, dzdt])
```

```python
# i
xa = np.array([1, 0.5, 0.1]) # Condiciones iniciales
p = (10, 0.5, 8/3) # parámetros
a, b = 0, 10 # Intervalo de tiempo
h = 0.01 # Paso de integración
k1 = (b-a)/h # Número de pasos
k1 = int(k1)
t1, w1 = integrador_ode(rk4, lorenz, xa, a, b, k1, p) # Integración
```

```python
k1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(w1[0], w1[1], w1[2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
